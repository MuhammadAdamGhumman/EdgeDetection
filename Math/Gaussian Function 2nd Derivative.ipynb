```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

sigma = 0.5

fig = plt.figure()
ax = plt.axes(projection="3d")
def z_function(x, y):
    return (((x**2)-(sigma**2))*(np.exp(-(x**2 + y**2)/(2*(sigma**2)))))/sigma**4

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#plt.show()

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_title('Gaussian Function 2nd Derivative');
plt.savefig('Gaussian Function 2nd Derivative.png', figsize=(10, 10), dpi=100, bbox_inches='tight')
```
